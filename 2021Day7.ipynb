```python
def realToFuel(x):
    curFuelPrice = 0
    for k in range(x+1):
        curFuelPrice = curFuelPrice + k
    return curFuelPrice
```

```python
def testPos2(xs,xtry):
    cursum = 0
    for x in xs:
        realDist = np.abs(x-xtry)
        cursum = cursum + realToFuel(realDist)
        # cursum = cursum  + np.abs(x-xtry)
    return cursum
```

```python
# Improved solution for part 2, made after solving it

maxPos = max(inis)
fuelConv = [realToFuel(x) for x in np.arange(maxPos)]

def testPos2Fast(xs,xtry):
    cursum = 0
    for x in xs:
        realDist = np.abs(x-xtry)
        cursum = cursum + fuelConv[realDist-1]
        # cursum = cursum  + np.abs(x-xtry)
    return cursum
    
curMin = 99999999999
bestPos = -1
for x in range(max(inis)):
    newTest = testPos2Fast(inis,x)
    if newTest < curMin:
        curMin = newTest
        bestPos = x
s2 = curMin 
print(f'Solution to part 2: {s2}')
```
