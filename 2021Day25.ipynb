```python
def EastMove(curLine):
    newLine = curLine.copy()
    wi = len(curLine)
        

    for k in range(wi-1,-1,-1):
        thisS = curLine[k]
        # Get the next to try to move to (looping around)
        nextS = curLine[np.mod(k+1,wi)]
        
        if (thisS == '>'):
            if (nextS == '.'):
                newLine[np.mod(k+1,wi)] = '>'
                newLine[k] = '.'
    # print(*np.array(newLine))
    return(newLine)

def SouthMove(curCol):
    newCol = curCol.copy()
    he = len(curCol)
        

    for k in range(he-1,-1,-1):
        thisS = curCol[k]
        # Get the next to try to move to (looping around)
        nextS = curCol[np.mod(k+1,he)]
        # print(thisS,nextS)
        
        if (thisS == 'v'):
            if (nextS == '.'):
                newCol[np.mod(k+1,he)] = 'v'
                newCol[k] = '.'
    # print(*np.array(newCol))
    return(newCol)
```

```python
def oneMove(curState):
    he = len(curState)
    wi = len(curState[0])
    newState = []
    
    # East moves
    for y in range(he):
        curLine = list(curState[y])
        newState.append(EastMove(curLine))
    
    # South moves
    for x in range(wi):
        curCol = []
        for y in range(he):
            curCol.append(newState[y][x])
        newCol = SouthMove(curCol)
        for y in range(he):
            newState[y][x] = newCol[y]
            
    return newState
    
# oneMove(oneMove(curState))
```
