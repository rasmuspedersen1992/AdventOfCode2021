```python
# Part 1 
openChars = ['(','[','{','<']

def openRec(line,pos):
    if len(line)== 0:
        return 'Legal'
    preL = line[:pos]
    postL = line[pos+1:]
    toBeClosed = line[pos]
    prevClosed = line[pos-1]


    # print(toBeClosed)
    # print(preL)
    # print(postL)
    # print(f'Line: {preL} --- {toBeClosed} --- {postL}')
    
    # print(f'cur post: {len(postL)}')
    if (len(postL) == 0):
        return 'Incomplete'

    # If next is open, go further down
    if toBeClosed in openChars:
        # print(line[pos:])
        # toReturn = openRec(line[1:],line[0])
        # toReturn = openRec(postL,0)
        toReturn = openRec(line,pos+1)
        return toReturn
    
    # print(prevClosed)
    if (prevClosed == '('):
        if (toBeClosed == ')'):
            # print('Closed ( and ) pair')
            # print(f'{preL[:-1]} --- {prevClosed}{toBeClosed} --- {postL}')
            # print(f'{preL[:-1]} ---    --- {postL}')
            newLine = preL[:-1] + postL
            toReturn = openRec(newLine,pos-1)
            return toReturn
            
    if (prevClosed == '['):
        if (toBeClosed== ']'):
            # print('Closed [ and ] pair')
            # print(f'{preL[:-1]} --- {prevClosed}{toBeClosed} --- {postL}')
            # print(f'{preL[:-1]} ---    --- {postL}')
            newLine = preL[:-1] + postL
            toReturn = openRec(newLine,pos-1)
            return toReturn
            
            
    
    if (prevClosed == '{'):
        if (toBeClosed== '}'):
            # print('Closed ( and } pair')
            # print(f'{preL[:-1]} --- {prevClosed}{toBeClosed} --- {postL}')
            # print(f'{preL[:-1]} ---    --- {postL}')
            newLine = preL[:-1] + postL
            toReturn = openRec(newLine,pos-1)
            return toReturn
            
            
    
    if (prevClosed == '<'):
        if (toBeClosed== '>'):
            # print('Closed < and > pair')
            # print(f'{preL[:-1]} --- {prevClosed}{toBeClosed} --- {postL}')
            # print(f'{preL[:-1]} ---    --- {postL}')
            newLine = preL[:-1] + postL
            toReturn = openRec(newLine,pos-1)
            return toReturn
    # If it reaches this point, line is corrupted
    return f'Corrupted. Next wrong sign is: {toBeClosed}'


# curL = inputs[0]
# curL = inputs[2]
# # print(curL)
# # curL = "[<>({}){}[([])<>]]"
# # curL = '<([{}])>'
# # openRec(curL,0)
# # curL
# openRec(curL,0)
```

```python
# Part 2
# Start by having only incomplete lines
newLines = []
for k in range(len(inputs)):
    curL = inputs[k]
    lineResult = openRec(inputs[k],0)

    if (lineResult[0] == 'I'): 
        newLines.append(curL)
# newLines
```

```python
# Part 2
def openRec2(line,pos):
    if len(line)== 0:
        return 'Legal'
    preL = line[:pos]
    postL = line[pos+1:]
    toBeClosed = line[pos]
    prevClosed = line[pos-1]


    # print(f'Line: {preL} --- {toBeClosed} --- {postL}')
    
    if (len(postL) == 0):
        # return 'Incomplete'
        # print(line)
        # print(toBeClosed)
        if toBeClosed in openChars:
            return line
        else:
            return preL[:-1]
        # return preL[:-1]

    # If next is open, go further down
    if toBeClosed in openChars:
        # toReturn = openRec2(line[1:],line[0])
        # toReturn = openRec2(postL,0)
        toReturn = openRec2(line,pos+1)
        return toReturn
    
    if (prevClosed == '('):
        if (toBeClosed == ')'):
            newLine = preL[:-1] + postL
            toReturn = openRec2(newLine,pos-1)
            return toReturn
            
    if (prevClosed == '['):
        if (toBeClosed== ']'):
            newLine = preL[:-1] + postL
            toReturn = openRec2(newLine,pos-1)
            return toReturn
            
            
    
    if (prevClosed == '{'):
        if (toBeClosed== '}'):
            newLine = preL[:-1] + postL
            toReturn = openRec2(newLine,pos-1)
            return toReturn
            
            
    
    if (prevClosed == '<'):
        if (toBeClosed== '>'):
            newLine = preL[:-1] + postL
            toReturn = openRec2(newLine,pos-1)
            return toReturn
    # If it reaches this point, line is corrupted
    return f'Corrupted. Next wrong sign is: {toBeClosed}'

# openRec2(newLines[1],0)
```

```python

def ReversePars(li):
    # print(li)
    toReturn = ''
    curScore = 0
    for k in range(len(li)):

        curScore = curScore * 5
        curOpen = li[-k-1]
        if (curOpen == '('):
            toReturn = toReturn + ')'
            curScore = curScore + 1
        if (curOpen == '['):
            toReturn = toReturn + ']'
            curScore = curScore + 2
        if (curOpen == '{'):
            toReturn = toReturn + '}'
            curScore = curScore + 3
        if (curOpen == '<'):
            toReturn = toReturn + '>'
            
            curScore = curScore + 4
    # return toReturn
    return curScore

# curL = newLines[0]
# asdf = openRec2(curL,0)
# ReversePars(asdf)
```

```python
allScores = []
for k in range(len(newLines)):
    # print(openRec(inputs[k],0))
    curL = inputs[k]
    curScore = ReversePars(openRec2(newLines[k],0))
    allScores.append(curScore)

sortedScores = np.sort(np.array(allScores))

s2 = sortedScores[int(np.floor(len(sortedScores)/2))]
print(f'Part 2: {s2}')
```
